```python
import numpy as np
import math

class NN():

    def __init__(self, input_size, units=1):

        w = np.random.randn(input_size, units) * 0.01
        b = np.random.randn(units)

        self.weight = w
        self.bias = b 
        self.dw = 0
        self.db = 0

    def linearLayer(self, X):

        self.X = X

        A_out = X @ self.weight + self.bias
        return A_out

    def Activation(self, A_out, activation=""):

        if activation == "sigmoid":
            out = 1 / (1 + np.exp(-A_out))
            return out

        if activation == "tanh":
            out = (np.exp(A_out) - np.exp(-A_out)) / (
                np.exp(A_out) + np.exp(-A_out)
            )
            return out
 
        A_out = np.maximum(0, A_out)

        return A_out

    def zero_grad(self):
        self.dw = 0
        self.db = 0

    def grad_update(self, outer_grad):

        # dW
        self.dw = self.X.T @ outer_grad

        # db
        self.db = np.sum(outer_grad, axis=0)

        # gradient for previous layer
        grad_prev = outer_grad @ self.weight.T

        return grad_prev
    
    def step(self, lr=0.001, optimizer=""):

        # update parameter for single layer
        # Default SGD
        self.weight = self.weight - lr * self.dw
        self.bias = self.bias - lr * self.db

        # print("dw norm:", np.linalg.norm(self.dw))


 
# nn_ = [layer_1, layer_2, layer_3] 

def backward(y, y_hat, nn_):

    m = y.shape[0]

    # dL/dy_hat for MSE
    outer_grad = (2 / m) * (y_hat - y)

    grad = outer_grad

    for layer in reversed(nn_):
        layer.zero_grad()
        grad = layer.grad_update(grad)

    return
```

```python
# Creating simple test dataset 

x_1 = np.arange(1, 100, 10)
x_2 = np.arange(650, 100, -55)
x_3 = x_1 * x_2
x_3 = (x_3 - x_3.min())/(x_3.max() - x_3.min()) + np.random.rand(10)
x_3 = [float(round(val, 2)) for val in x_3]
print(x_3)
# Y =  2*x1​ + 0.5*x2​ + 3*x3​
x = [x_1.tolist(), x_2.tolist(), x_3]
Y = 2*x_1 + 0.5*x_2 + 3*np.array(x_3)
Dataset = [x_1.tolist(), x_2.tolist(), x_3, Y.tolist()]
print("Dataset : ", Dataset)
X_train = Dataset[:-1]
y_train = Dataset[-1]

X_train = np.array(X_train)
y_train = np.array(y_train)
X_train = X_train.T
y_train = y_train.reshape(-1, 1)
```

```python
input_layer = NN(input_size=X_train.shape[1], units=4)  # input_size ~ no. of input feautures 
layer_1 = NN(input_size=4, units=4)
layer_2 = NN(input_size=4, units=4)
layer_3 = NN(input_size=4, units=4)
layer_4 = NN(input_size=4, units=1)
```

```python
X_train = (X_train - X_train.mean(axis=0))/X_train.var(axis=0)
y_train = (y_train - y_train.mean())/y_train.var()
print(y_train.shape)
```

```python
epochs = 100
input_layer = NN(input_size=X_train.shape[1], units=4)  # input_size ~ no. of input feautures 
layer_1 = NN(input_size=4, units=4)
layer_2 = NN(input_size=4, units=4)
layer_3 = NN(input_size=4, units=4)
layer_4 = NN(input_size=4, units=1)

def model(X): 
    X = input_layer.linearLayer(X)
    X = layer_1.linearLayer(X)
    X = layer_2.linearLayer(X)
    X = layer_3.linearLayer(X)
    Y_hat = layer_4.linearLayer(X)

    nn_ = [input_layer, layer_1, layer_2, layer_3, layer_4]
    return Y_hat, nn_


for i in range(epochs): 

    output, nn_ = model(X_train)
    backward(y_train, output, nn_)
    loss = np.mean((output - y_train)**2)
    # print(f"ytrain shape {y_train.shape}, output {output.shape}")
    print(f"Epoch {i} Completed || Current Loss : {loss}")
    # Y =  2*x1​ + 0.5*x2​ + 3*x3​
    # print("Predicted Value for x1  = 1, x2 = 1, x3 = 1", model([1, 1, 1]))
    for i, layer in enumerate(reversed(nn_)):
        layer.step()
        # print(f"layer {4-i}")

```
