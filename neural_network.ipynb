```python
import math
class NN(): 

    def __init__(self, X, input_size, units=1): 

        w = np.zeros((input_size, units)) 
        b = np.zeros(units)
        self.weight =  w 
        self.bias = b 
        self.X = X 
        self.dw = 0 
        self.db = 0 
        


    def linearLayer(self):
 
        A_out = np.dot(self.X, self.weight) + self.bias  # A_out = X @ W.T +  B

        return A_out
    
    def Activation(self, A_out, activation=""): 
        
        if activation == "sigmoid": 
            
            out = 1/(1 + np.exp(-A_out))
            
            return out
        if activation == "tanh": 

            out = (np.exp(A_out) - np.exp(-A_out))/(np.exp(A_out) + np.exp(-A_out))

            return out 
        
        
        for i in range(A_out.shape[0]):
            for j in range(A_out.shape[1]): 
                if A_out[i] < 0: 
                    A_out[i] = 0 
        
        return A_out 
    
    def zero_grad(self): 
        self.dw = 0 
        self.db = 0 

    def grad_update(self, outer_grad):
        
        self.dw = self.X.T @ outer_grad  # like prev_input * W2 * W2 * W3  
        self.bw = outer_grad
        outer_grad = self @ outer_grad

        return self.dw, self.db 
    
    


        


```
