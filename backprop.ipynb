```python
class Value: 

    def __init__(self, data, children=(), op=''): 
        self.data = data
        self.children = children
        self.op = op


    def __add__(self, other): 
        out = Value(self.data  + other.data, children=(self, other), op='+') 
        return out 
    
    def __mul__(self, other): 
        out = Value(self.data  * other.data, children=(self, other), op='*') 
        return out 

    def __repr__(self):
        return f"Value : {self.data}, children node : {self.children}, operator : {self.op}"
        
    
```

```python
a = Value(3) ; b  = Value(4)
```
